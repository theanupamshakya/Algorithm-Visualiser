```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv('dataset.csv')
```

```python
# df.head()
```

```python
data = {
    "Name": ['Anupam', 'User1', 'User2', 'User3', 'User4'],
    "Age": [24, 25, 26, 27, 28],
    "Salary": [50000, 20000, 12500, 89000, 98000]
}
```

```python
df = pd.DataFrame(data)
```

```python
print("Basic Dataset: ")
print(df.head())
```

```python
print("\nBasic Statistics: ")
print(df.describe())
```

```python
# mean salary
mean_salary = df["Salary"].mean()
print("\nAverage Salary:", mean_salary)
```

```python
# histogram
plt.hist(df['Salary'], bins=5, color='red', edgecolor='black')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution')
plt.show()
```

```python
# plot
plt.plot(df['Name'], df['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary by Name')
plt.show()
```

```python
# bar
plt.bar(df['Name'], df['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary by Name')
plt.show()
```

```python
# grid
plt.grid(data)
plt.plot(df['Name'], df['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary by Name')
plt.show()
```

```python
# sub plot
plt.subplot(1, 2, 1)
plt.plot(df['Name'], df['Salary'])
plt.xlabel('Name')
plt.ylabel('Salary')
plt.title('Salary by Name')
plt.show()
```

```python
# scatter plot
plt.scatter(df['Age'], df['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary by Age")
plt.show()
```
